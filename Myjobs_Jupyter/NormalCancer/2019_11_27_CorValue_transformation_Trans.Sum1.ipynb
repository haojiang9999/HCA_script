```r
##### Loda prepared data
Log2.Cor.Res.list <- readRDS("/data8t_4/JH/MyJobs/NormalCancer_scColon/Log2.Cor.Res.list.dataset.rds")
```

```r
#### Unpack the dataset
Cor.Res.CV1000 <- Log2.Cor.Res.list$Cor.Res.CV1000
Cor.Res.CV1500 <- Log2.Cor.Res.list$Cor.Res.CV1500
Cor.Res.CV2000 <- Log2.Cor.Res.list$Cor.Res.CV2000
Cor.Res.CV4000 <- Log2.Cor.Res.list$Cor.Res.CV4000
Cor.Res.CV8000 <- Log2.Cor.Res.list$Cor.Res.CV8000
Pheno.merged <- Log2.Cor.Res.list$Pheno.merged
```

```r
#### Raw Correlation value transformed into correlation percent for each cell
#rawCorrelation[1:5,1:5]
Trans.sum1.1000<- base::apply(Cor.Res.CV1000$Cor.merged, 2, function(x){
  x/sum(x)
})
Trans.sum1.1500<- base::apply(Cor.Res.CV1500$Cor.merged, 2, function(x){
  x/sum(x)
})
Trans.sum1.2000<- base::apply(Cor.Res.CV2000$Cor.merged, 2, function(x){
  x/sum(x)
})
Trans.sum1.4000<- base::apply(Cor.Res.CV4000$Cor.merged, 2, function(x){
  x/sum(x)
})
Trans.sum1.8000<- base::apply(Cor.Res.CV8000$Cor.merged, 2, function(x){
  x/sum(x)
})
```

```r
source("/data8t_4/JH/MyJobs/1_R_script/R_Plot/DimeReduPlot.R")
DimeReduPlot(mx = Trans.sum1.1000, color = Pheno.merged$Cell_info, 
             tiltle = "Cor.Res.CV1000_Cell_info", print = T, perplexity = 50,check_duplicates = FALSE)
```
